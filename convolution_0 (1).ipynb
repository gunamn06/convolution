```python
import torch

X = torch.tensor(X, dtype=torch.float32)   
Y = torch.tensor(Y, dtype=torch.long)      
```

```python
from torch.utils.data import TensorDataset, DataLoader

dataset = TensorDataset(X, Y)
batches = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
X = torch.FloatTensor(X)
Y = torch.LongTensor(Y)
```

```python
X.shape
```

```python
model = nn.Sequential(
    nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1),
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=2, stride=2),
    nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=2, stride=2),
    nn.Flatten(),
    nn.Linear(64*2*2, 128),
    nn.ReLU(),
    nn.Linear(128, 3)
)
```

```python
loss = nn.CrossEntropyLoss()
opt = Adam(model.parameters(), lr=0.001)
```

```python
import torch
import torch.nn as nn

lossfn = nn.CrossEntropyLoss()   # define loss
opt = torch.optim.SGD(model.parameters(), lr=0.01)

for xb, yb in batches:
    opt.zero_grad()
    
    yp = model(xb)
    
    loss = lossfn(yp, yb)   # now it works
    
    loss.backward()
    opt.step()
```

```python
class MyNet(nn.Module):
  def __init__(self):
    super().__init__()
    self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
    self.maxm = nn.MaxPool2d(kernel_size=2, stride=2)
    self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
    self.relu = nn.ReLU()
    self.flatten = nn.Flatten()
    self.linear1 = nn.Linear(64*2*2, 128)
    self.linear2 = nn.Linear(128, 3)

  def forward(self, x):
    x = self.conv1(x)
    x = self.relu(x)
    x = self.maxm(x)
    x = self.conv2(x)
    x = self.relu(x)
    x = self.maxm(x)
    x = self.flatten(x)
    x = self.linear1(x)
    x = self.relu(x)
    x = self.linear2(x)
    return x
     
```

```python
from torch.utils.data import TensorDataset, DataLoader
import torch.nn as nn

# create dataset and dataloader
dataset = TensorDataset(X, Y)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

lossfn = nn.CrossEntropyLoss()

model.train()
c = 0

for epoch in range(100):
    c += 1
    epoch_loss = 0.0
    
    for batch_X, batch_Y in dataloader:
        opt.zero_grad()
        
        yp = model(batch_X)
        
        ls = lossfn(yp, batch_Y)
        
        ls.backward()
        opt.step()
        
        epoch_loss += ls.item()

    if c % 10 == 0:
        print(epoch_loss)
```
